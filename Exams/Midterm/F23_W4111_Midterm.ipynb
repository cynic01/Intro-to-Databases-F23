Provide a short (two or three sentence) definition/description of the following terms. Some of these concepts do not have a single, precise, agreed definition. You may find slight differences in your research. Focus on the concept and grading will be flexible.

1. Super Key<br><br>

1. Primary Key<br><br>
5. Alternate Key<br><br>

2. Natural Key<br><br>
2. Surrogate Key<br><br>
2. Substitute Key<br><br>
7. Foreign Key<br><br>
8. External Key

<u>Answer</u>

1. A Super Key is a set of one or more attributes (columns) that can uniquely identify a row in a table. For example, a super key can be a combination of attributes "student_id" and "email" in a Student database.

2. A candidate key is a minimal super key, which means that if we remove any attribute (column) from a candidate key, there would be duplicate combinations of values in the remaining attributes (columns).

3. A primary key is one of all possible candidate keys that is chosen to index into the table and selecting single rows. Also, it cannot allow null values.

4. Alternate keys are candidate keys that are not chosen to be the primary key.

5. A unique key can uniquely identify a single row in a table in order to prevent duplicates. A unique key can be defined to allow nulls.

6. A natural key is made from data that exists outside the current database (not generated by the system), such as a social security number.

7. A surrogate key is an artificial key made from data that is generated by the system, and are usually numeric values.

8. A substitute key is the same as an alternate key (candidate keys not chosen to be the primary key).

9. A foreign key is a set of attributes (columns) that refers to the primary key of another table.

10. An external key is the same as a foreign key (primary key of another table).

- Self-join

<u>Answer</u>

1. A natural join is a join operation that implicitly creates the join clause using common column names in the two tables. It is convenient when the two tables you want to join has matching names.

## Birth Countries and Death Countries

- __Note:__ Use the instructions from HW1 Part 2 Nonprogramming to load the Lahman's Baseball Database data into MySQL if you have not already done so.<br><br>

- In Lahman's Baseball Database table `people`, there is information about people's `birthCountry` and `deathCountry`.<br><br>

- There are countries in which at least one person was born but in which no person has died.<br><br>

- Write a query that produces a table of the form:
    - `birthCountry`
    - `no_of_births`, which is the total number of births in the country.<br><br>

- The table contains all rows for countries in which there with births but no deaths.<br><br>

- You may create indexes to improve performance.

<u>Answer</u>


````python
%%sql

USE lahmans_hw1;

WITH Births(birthCountry, no_of_births) AS
    (SELECT birthCountry, COUNT(*) FROM People GROUP BY birthCountry),
    Deaths(deathCountry, no_of_deaths) AS
    (SELECT deathCountry, COUNT(deathCountry) FROM People GROUP BY deathCountry HAVING COUNT(deathCountry) > 0)
    SELECT Births.birthCountry, Births.no_of_births
        FROM Births LEFT JOIN Deaths ON Births.birthCountry = Deaths.deathCountry
        WHERE no_of_deaths IS NULL
        ORDER BY no_of_births DESC;
````

<u>Answer</u>


````python
%%sql

USE lahmans_hw1;

WITH Bat(playerID, great_because, slg, obp, obp_slg, total_abs) AS (
    SELECT playerID,
        'Batter' AS great_because,
        (SUM(h) - SUM(`2b`) - SUM(`3b`) - SUM(hr) + 2*SUM(`2b`) + 3*SUM(`3b`) + 4*SUM(hr)) / SUM(ab),
        (SUM(h) + SUM(BB)) / (SUM(ab) + SUM(BB)),
        (SUM(h) + SUM(BB)) / (SUM(ab) + SUM(BB)) + (SUM(h) - SUM(`2b`) - SUM(`3b`) - SUM(hr) + 2*SUM(`2b`) + 3*SUM(`3b`) + 4*SUM(hr)) / SUM(ab) AS obp_slg,
        SUM(ab) AS total_abs
    FROM Batting GROUP BY playerID
    HAVING total_abs >= 3000 AND obp_slg >= 1.000),
Pitch(playerID, great_because, total_w, total_l, total_d, win_percentage) AS (
    SELECT playerID,
        'Pitcher' AS great_because,
        SUM(w) AS total_w,
        SUM(l),
        SUM(w) + SUM(l) AS total_d,
        SUM(w) / (SUM(w) + SUM(l)) AS win_percentage
    FROM Pitching GROUP BY playerID
    HAVING total_d >= 200 AND win_percentage >= 0.7 OR total_w >= 300),  -- Can change to win_percentage >= 0.666 to match answer above
Final(playerID, great_because, slg, obp, obp_slg, total_abs, total_w, total_l, total_d, win_percentage) AS (
    SELECT playerID, great_because, slg, obp, obp_slg, total_abs, NULL AS total_w, NULL AS total_l, NULL AS total_d, NULL AS win_percentage FROM Bat
    UNION ALL
    SELECT playerID, great_because, NULL AS slg, NULL AS obp, NULL AS obp_slg, NULL AS total_abs, total_w, total_l, total_d, win_percentage FROM Pitch)
SELECT Final.playerID AS playerID, nameLast, nameFirst, great_because, slg, obp, obp_slg, total_abs, total_w, total_l, total_d, win_percentage
FROM Final NATURAL JOIN People
ORDER BY playerID;
````
