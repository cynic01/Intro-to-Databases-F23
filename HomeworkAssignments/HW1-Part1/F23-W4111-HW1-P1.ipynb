1. What is a database management system and how do relational databases organize data?

A DBMS is a piece of software that gives users the ability to store, retrieve, and manipulate data in a database. Relational databases organize data by tables, which are connected by Primary Keys and Foreign Keys.

2. Columbia University uses several applications that use databases to run the university. Examples are SSOL and CourseWorks. An alternate approach could be letting students, faculty, admimistrators, etc. use shared Google Sheets to create, retrieve, update and delete information. What are some problems with the shared sheet approach and what functions do DMBS implement to solve the problems?

A problem would be that Google Sheets lack strong data integrity enforcement, and anyone with access can modify data, leading to errors or inconsistencies. DBMS implements data integrity through constraints and validation rules, so accuracy and consistency is ensured. Another problem is that Google Sheets has limited access control over who can view, edit and delete each piece of data, while DBMS can provide fine-grained access control to protect sensitive information.

3. Explain the differences between SQL, MySQL Server and DataGrip.

SQL is a language used to interact with databases, MySQL Server is a DBMS that uses SQL, and DataGrip is an Integrated Development Environment (IDE) that provides an interface to work with database systems like MySQL Server.

4. What are 4 different types of DBMS table relationships. Give a brief explanaition for each.

With the Crow's Foot notation, the relationship types are One or Zero (where an entry of A is related to 0 or 1 entries of B), One and only One (where an entry of A is related to exactly one entry of B), Zero or Many (where an entry of A is related to any number of entries of B), and One or Many (where an entry of A is related to one or more entries of B).

5. What is a primary key and why is it important?

A primary key is a key that occurs only once in the table. It is important as we use it to index the table and build dependency for foreign keys in other tables.

6. The relational algebra is closed under the operators. Explain what this means and give an example.

This means that when we use relational algebra to perform operations on existing relations (tables), the results are also relations (tables). For example, when we use projection (π) on the relation Classroom with attributes building, room_number, and capacity to select only the attribute building, the result is also a relation with a tabular structure.

7. SQL is a declarative data manipulation language. What are some pros and cons of declarative DMLs relative to procedural ones?

Delarative DMLs have pros such as simplicity to use (describe what we want instead of step-by-step procedure), optimization (DBMS can choose most efficient way to execute query), and abstraction (we don't need to understand details of data access and manipulation). They also have cons such as debugging (can be hard to find out what is wrong about query), limited control (the DBMS instead of us decideds how to execute queries), and hard to express complex queries (when doing it using a procedural DML is simpler to think about).

8. Some of the Columbia University databases/applications represent the year/semester attribute of a section in the form "2023_2". The first four characters are the academic year, and the last character is the semester (1, 2, or 3). The data type for this attribute might be char(6). Using this example, explain the concepts of domain and atomic domain. How is domain different from type?

Domain is the set of values that a particular attribute can take. In the example it is "yyyy_S" where yyyy is the set of 4 characters representing the academic year and S is the set of one character (1, 2, or 3) representing the semester. Atomic domain is the smallest indivisible unit of a domain. In this example the academic year is an atomic domain, and the semester is another atomic domain. Domain is different from data type in that domain is defined by the database creator / administrator to restrict the possible values in an attribute, while data type is defined by the DBMS and used to represent the values of the data.

9. What is the semi-structured data model and how is it different from the relational data model?

Semi-structured data model is a more flexible and less rigid data representation model than relational data model. In a semi-structured data model, the structure doesn't have to be predefined and instead is shaped however the application finds most convenient, complex nested data structures are allowed, and querying may not use SQL at all.

10.  Briefly explain the difference between a schema and instance.

A schema is a framework that defines how data should be stored, e.g. tables, columns, data types, constraints. An instance contains the actual data adhering to the schema in a particular point in time.

## SQL 2

Write a query that returns the names and departments of instructors who advised students that got an A in some course. Your answer should be sorted by name and should not contain duplicates.

You may use the `takes`, `student`, `instructor`, and `advisor` tables. **You may not use the `IN` operator.**

````python
%%sql

select distinct instructor.name, instructor.dept_name
    from (takes left join advisor on takes.ID = advisor.s_id)
        left join instructor on i_ID = instructor.ID
    where grade = 'A'
    order by name;
````
